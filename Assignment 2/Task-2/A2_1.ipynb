```python
import json

dataset = {}

with open(r"DATASET_TASK2\task2_test.json", "r") as data_file:
    data = json.load(data_file)
    for i in range(len(data)):
        text = data[i]['raw_words']
        labels = ['O'] * len(data[i]['words'])
        for aspects in data[i]['aspects']:
            start = aspects['from']
            end = aspects['to']

            labels[start] = "B"

            for j in range(start + 1, end):
                labels[j] = "I"

        dataset[i] = {
            "text": text,
            "labels": labels
        }

with open("ATE_test.json", "w") as outfile:
    json.dump(dataset, outfile, indent=4)


dataset = {}

with open(r"DATASET_TASK2\task2_train.json", "r") as data_file:
    data = json.load(data_file)
    for i in range(len(data)):
        text = data[i]['raw_words']
        labels = ['O'] * len(data[i]['words'])
        for aspects in data[i]['aspects']:
            start = aspects['from']
            end = aspects['to']

            labels[start] = "B"

            for j in range(start + 1, end):
                labels[j] = "I"

        dataset[i] = {
            "text": text,
            "labels": labels
        }

with open("ATE_train.json", "w") as outfile:
    json.dump(dataset, outfile, indent=4)

dataset = {}

with open(r"DATASET_TASK2\task2_val.json", "r") as data_file:
    data = json.load(data_file)
    for i in range(len(data)):
        text = data[i]['raw_words']
        labels = ['O'] * len(data[i]['words'])
        for aspects in data[i]['aspects']:
            start = aspects['from']
            end = aspects['to']

            labels[start] = "B"

            for j in range(start + 1, end):
                labels[j] = "I"

        dataset[i] = {
            "text": text,
            "labels": labels
        }

with open("ATE_val.json", "w") as outfile:
    json.dump(dataset, outfile, indent=4)

```
